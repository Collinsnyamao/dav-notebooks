```python
mens_perfumes = perfume_data.loc[perfume_data['department'] == 'Men']
mens_perfumes
```

```python
women_perfumes = perfume_data.loc[perfume_data['department'] == 'Women']
women_perfumes
```

```python
women_scent_popularity = women_perfumes.groupby('scents')['item_rating'].agg([
    'mean',
    'count'
]).reset_index()
popular_scents = women_scent_popularity[women_scent_popularity['count'] >= 5].sort_values('mean', ascending=False)
popular_scents.iloc[0]
```

```python
mens_scent_popularity = mens_perfumes.groupby('scents')['item_rating'].agg([
    'mean',
    'count'
]).reset_index()
popular_scents = mens_scent_popularity[mens_scent_popularity['count'] >= 5].sort_values('mean', ascending=False)
popular_scents.iloc[0]
```

```python
brand_ratings = perfume_data.groupby('brand')['item_rating'].agg(['mean', 'count'])
top_brands = brand_ratings[brand_ratings['count'] >= 5].sort_values('mean', ascending=False).head(10)
print("Top 10 Brands by Rating (min 5 products):")
print(top_brands)
```

```python
perfume_data.columns
```

```python
avg_rating = perfume_data.loc[(perfume_data['ml'] == 100) & (perfume_data['department'] == 'Women')]['new_price'].mean().round(4)
print(f"Average rating: {avg_rating:.2f}")
```

```python
seller_stats = perfume_data.groupby('seller').agg({
    'name': 'count',
    'item_rating': 'mean',
    'num_seller_ratings': 'first'
}).rename(columns={'name': 'product_count'}).sort_values('product_count', ascending=False)

print("Top sellers:")
print(seller_stats.head(10))
```
