## Summary


```python
houses_cleaned = houses_data_no_id
```

```python
none_features = ['Alley', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                'BsmtFinType2', 'FireplaceQu', 'GarageType', 'GarageFinish',
                'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']
none_features
```

```python
for feature in none_features:
    if feature in houses_cleaned.columns:
        houses_cleaned[feature] = houses_cleaned[feature].fillna('None')
```

```python
houses_cleaned['MasVnrType'] = houses_cleaned['MasVnrType'].fillna('None')
houses_cleaned['MasVnrArea'] = houses_cleaned['MasVnrArea'].fillna(0)
```

```python
houses_cleaned['LotFrontage'] = houses_cleaned.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)

#used the median of the neighbors
houses_cleaned['LotFrontage']
```

```python
houses_cleaned['GarageYrBlt'] = houses_cleaned['GarageYrBlt'].fillna(houses_cleaned['YearBuilt'])

#used the YearBuilt field
```

```python
electrical_mode = houses_cleaned['Electrical'].mode()[0]
houses_cleaned['Electrical'] = houses_cleaned['Electrical'].fillna(electrical_mode)

#used the most common value
```
